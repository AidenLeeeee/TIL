# 1.2. 변수명

- **{{ 변수명 }}**

### **`test.py`**

```python
from flask import Flask, render_template
app = Flask(__name__)


@app.route('/hello/<user>')
def hello_name(user):
    return render_template('variable.html', name=user)


if __name__ == '__main__':
    app.run(host="0.0.0.0", port="8080")
```

- **render_template** 과 함께 **user** 변수를 **name** 에 담아 전달

### **`variable.html`**

- **variable.html** 파일에서는 **test.py** 에서 전달받은 **name** 변수를 받아 처리
- **test.py** 에서 여러 개의 변수를 전달하면 다양한 변수를 활용 가능

# 1.3. 반복문

- **{% for i in list %}**<br>
  **{{ value }}**<br>
  **{% endfor %}**
<br>
<br>
- **for** 로 선언 후, **endfor** 로 종료
<br>
<br>
- **indentation** 은 사용하지 않음
<br>
<br>
- **range()**: 사용 가능
    - **range(len(list))**: 사용 불가
    - **len(list)**: list | length 로 사용 가능
- **loop.index**: 반복문 횟수를 반환 가능
- **list[index]**: 인덱스 접근 가능

### **`loop_test.py`**

```python
from flask import Flask, render_template
app = Flask(__name__)

@app.route('/hello_loop')
def hello_name():
    value_list = ['list1', 'list2', 'list3']
    return render_template('loop.html', values=value_list)

if __name__ == '__main__':
    app.run(host="0.0.0.0", port="8080")
```

- **render_template** 과 함께 **values** 에 리스트를 담아 전달

### **`loop.html`**

- 마찬가지로, **loop.html** 에서는 전달받은 리스트 요소들을 **for 문** 을 돌며 화면에 리스트 아이템으로 랜더링

# 1.4. 조건문

- **{% if %} {% elif %} {% else %} {% endif %}**
<br>
<br>
- **Indentation** 은 사용하지 않음
<br>
<br>
- **{% if data >= 30 %}**<br>
  **\<h3>30 보다 큽니다.\</h3>**<br>
  **{% elif data > 25 %}**<br>
  **\<h3>25 보다 큽니다.\</h3>**<br>
  **{% else %}**<br>
  **\<h3>{{ data }}\</h3>**<br>
  **{% endif %}**

### **`condition_test.py`**

```python
from flask import Flask, render_template
app = Flask(__name__)

@app.route('/hello_if')
def hello_html():
    value = 27
    return render_template('condition.html', data=value)

if __name__ == "__main__":
    app.run(host="0.0.0.0", port="8080")
```

- **render_template** 과 함께 **data** 변수를 전달

### **`condition.html`**

- **Jinja2** 의 조건문 문법을 활용해 전달받은 **data** 를 처리
