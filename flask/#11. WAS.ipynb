# 1.1. Web Server vs WSGI

### **`Web Server`**

- 웹서버는 정적인 Html 페이지를 반환
    - Apache 와 Nginx...

### **`WAS`**

- 웹서버 위에서 동작하는 서버 응용 프로그램
<br>
<br>
- 웹서버가 동적으로 데이터를 반환하도록 하기 위해서는 WAS 프레임워크가 필요
    - 주요 WAS 로는 Flask, Django, rails, node.js...
    - 이러한 프레임워크를 사용해서 WAS 개발

### **`WSGI`**

- Web Server Gateway Interface
<br>
<br>
- Apache, Nginx 와 같은 **웹서버**와 Flask, Django 와 같은 **WAS** 가 통신할 수 있도록 하는 프로토콜
<br>
<br>
- WSGI 는 Python 응용 프로그램이 웹 서버와 통신할 수 있도록 하는 프로토콜

- 주요 기능
    - 웹서버의 환경 정보와 콜백 함수를 Python 응용 프로그램에 전달
    - Python 응용 프로그램에서 요청을 처리한 결과를 콜백 함수를 통해 웹서버에 전달

### **`Middleware`**

- WSGI 의 동작을 지원하는 프로그램
<br>
<br>
- 주요 기능
    - Middleware 가 응용 프로그램을 실행
    - 웹서버의 환경 정보 변경 시 이에 따른 요청 경로 변경
    - 동일 프로세스 상에서 여러 응용 프로그램 실행 지원
    - XSLT 스타일 시트 적용 등 전처리 지원
<br>
<br>
- 주요 Middleware
    - mod_wsgi, uwsgi, gunicorn, twisted.web, tornado ...

# 1.2. Summary

- 정적인 파일을 제공하는 **웹서버**, 동적인 데이터 처리를 수행하여 동적인 파일 반환을 담당하는 **WAS** 가 존재
<br>
<br>
- 웹서버와 WAS 가 서로 통신하기 위해서는 프로토콜이 필요하고, **WSGI** 를 활용해 이를 지원한다.
<br>
<br>
- 이러한 WSGI 의 동작을 지원하기 위해 **Middleware** 가 필요하다.

- 즉, 웹서버와 파이썬 응용 프로그램의 통신을 지원하기 위해 WSGI 프로토콜을 사용하며, 이에 편의성 및 기능을 추가한 WSGI Middleware 가 존재

# 1.3. 동작 순서

- **웹서버**가 설정에 따라 특정 Port 를 Listening
<br>
<br>
- 해당 Port 로 요청이 들어오면 웹서버가 thread 생성, 요청 처리
    - 이 때, thread 는 **WSGI Middleware** 를 통해 **Python 응용 프로그램(WAS)**에 요청 전달
    - **Python 응용 프로그램(WAS)**이 요청 처리한 후, 결과를 다시 **WSGI Middleware** 에 전달
<br>
<br>
- **WSGI Middleware** 가 **웹서버** thread 에 요청 처리 결과 전달
